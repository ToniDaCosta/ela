```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\ateso\Desktop\Flash\orders.csv")
```

```python
df.head()
```

```python
new_orders = pd.DataFrame()

new_orders['order_id'] = df['33557880']
new_orders['activation_time_local'] = df['2019-03-10 23:59:59.000000']
new_orders['country_code'] = df['AR']
new_orders['store_address'] = df['14200']
new_orders['final_status'] = df['DeliveredStatus']
new_orders['payment_status'] = df['PAID']
new_orders['products'] = df['1']
new_orders['products_total'] = df['4.54']
new_orders['purchase_total_price'] = df['8.64']
```

```python
new_orders['authorization_status'] = 'Correctly Authorized'
new_orders.loc[new_orders['products_total'] < new_orders['purchase_total_price'], 'authorization_status'] = 'Under-Authorized'

# Calculate the original percentage of under-authorized orders
original_under_authorized_percentage = (new_orders['authorization_status'] == 'Under-Authorized').mean()

# Incremental authorization thresholds (adjust as needed)
authorization_thresholds = [1.0]

# Simulate the impact of incremental authorization
for threshold in authorization_thresholds:
    simulated_correctly_authorized = (new_orders['products_total'] * threshold >= new_orders['purchase_total_price'])
    simulated_correctly_authorized_percentage = simulated_correctly_authorized.mean()
    print(f"With {threshold * 100:.2f}% incremental authorization, the percentage of correctly authorized orders is: {simulated_correctly_authorized_percentage:.2f}")
```

```python
new_orders['authorization_status'] = 'Correctly Authorized'
new_orders.loc[new_orders['products_total'] < new_orders['purchase_total_price'], 'authorization_status'] = 'Under-Authorized'

# Calculate the original percentage of under-authorized orders
original_under_authorized_percentage = (new_orders['authorization_status'] == 'Under-Authorized').mean()

# Simulate the impact of a 20% increase in the authorized amount
simulated_correctly_authorized = (1.2 * new_orders['products_total'] >= new_orders['purchase_total_price'])
simulated_correctly_authorized_percentage = simulated_correctly_authorized.mean()

print(f"With a 20% increase in the authorized amount, the percentage of correctly authorized orders is: {simulated_correctly_authorized_percentage:.2f}")
```

```python
new_orders['authorization_status'] = 'Correctly Authorized'
new_orders.loc[new_orders['products_total'] < new_orders['purchase_total_price'], 'authorization_status'] = 'Under-Authorized'

country_under_authorized_percentage = new_orders.groupby('country_code')['authorization_status'].apply(lambda x: (x == 'Under-Authorized').mean() * 100)

country_correctly_authorized_percentage = new_orders.groupby('country_code')['authorization_status'].apply(lambda x: (x == 'Correctly Authorized').mean() * 100)

print("Percentage of Under-Authorized Orders by Country:")
print(country_under_authorized_percentage)

print("\nPercentage of Correctly Authorized Orders by Country:")
print(country_correctly_authorized_percentage)
```

```python
new_orders['remaining_amount'] = new_orders['products_total'] - new_orders['purchase_total_price']

remaining_amount_by_store = new_orders.groupby('store_address')['remaining_amount'].sum().abs().reset_index()

most_difference_stores = remaining_amount_by_store.sort_values(by='remaining_amount', ascending=False)

most_difference_stores
```

```python
new_orders['remaining_amount'] = new_orders['products_total'] - new_orders['purchase_total_price']

remaining_amount_by_store = new_orders.groupby(['country_code'])['remaining_amount'].sum().abs().reset_index()

most_difference_stores = remaining_amount_by_store.sort_values(by='remaining_amount', ascending=False)

most_difference_stores_with_country = most_difference_stores.copy()

most_difference_stores_with_country

```

```python
new_orders['remaining_amount'] = new_orders['products_total'] - new_orders['purchase_total_price']

remaining_amount_by_country = new_orders.groupby(['country_code'])['remaining_amount'].sum().abs().reset_index()

most_difference_countries = remaining_amount_by_country.sort_values(by='remaining_amount', ascending=False).head(10)

most_difference_countries_with_country = most_difference_countries.copy()
print(most_difference_countries_with_country)


plt.figure(figsize=(10, 6))
sns.barplot(x='remaining_amount', y='country_code', data=most_difference_countries_with_country, color='lightgreen')
plt.xlabel('Remaining Amount')
plt.ylabel('Country Code')
plt.title('Top 10 Countries with Most Significant Differences in Remaining Amounts')
plt.show()

```

```python
under_auth_orders = new_orders[new_orders['authorization_status'] == 'Under-Authorized']

under_auth_orders.loc[:, 'final_status_binary'] = (under_auth_orders['final_status'] == 'DeliveredStatus').astype(int)

correlation_coefficient = under_auth_orders['remaining_amount'].corr(under_auth_orders['final_status_binary'])
correlation_coefficient

```
